# MILESTONE 1

a) Access to the website [BACHELOR IN DATA SCIENCE AND ENGINEERING
](https://www.uc3m.es/bachelor-degree/data-science)

b) Find the element tag with `id="program"` and print the result

c) Find the table inside PROGRAM for Course 1 - Semester 1 and print the result


```python
# set options to be headless
from selenium import webdriver
from selenium.webdriver.common.by import By

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
# open it, go to a website, and get results
driver = webdriver.Chrome(options=options)
```

```python
## a) Access to the website
url = "https://www.uc3m.es/bachelor-degree/data-science"

# 'Get' is used to take an action on browser: navigate to the web page.
page = driver.get(url)
```

```python
## b) Find the element tag with id="program" and print the result
item = driver.find_element(By.ID, value = "program")
print('The element with id "program" is:', item)
print('The text of the element is:', item.text)
```

```python
## c) Find the table inside PROGRAM for Course 1 - Semester 1 and print the result
container = driver.find_element(By.XPATH, "//*[contains(text(), 'Year 1 - Semester 1')]/ancestor::div")
table_element = container.find_element(By.TAG_NAME, "table")
print(table_element.text)
print('container',container.text)
```

# MILESTONE 2

a) Obtain the link to Web Analytics course by finding the corresponding element in the source code.

b) Access to this URL by clicking in the link.

c) Print the text inside the _Learning activities and methodology_ section.

```python
## a) Obtain the link to Web Analytics course by finding the corresponding element in the source code
container = driver.find_element(By.XPATH, "//*[contains(text(), 'Web Analytics')]/parent::a")
link = container.get_attribute('href')
print(link)
```

```python
## b) Access to this URL by clicking in the link
page = driver.get(link)
```

```python
## c) Print the text inside the Learning activities and methodology section.
container = driver.find_element(By.XPATH, "//*[contains(text(), 'Learning activities and methodology')]")
section = container.find_element(By.XPATH, "following-sibling::*")
print(section.text)
```
